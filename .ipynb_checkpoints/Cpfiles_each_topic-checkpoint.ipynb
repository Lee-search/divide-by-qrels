```python
with open('./filelist', 'r') as l:
    list_of_path = l.readlines()
list_of_id = []

temp = []
for path in list_of_path:
    list_of_id.append(path[-16:-5])
    temp.append(path[1:-1])    
list_of_path = temp

# list_of_id: NCT_id
# list_of_path: 실제경로
print(f"NCT_id:  {len(list_of_id)}")
print(f"실제경로: {len(list_of_path)}")
```

```python
## TREC Variable
with open('./cds-files/topics2021.xml', 'r', encoding='UTF8') as f:
    root = ET.parse(f).getroot()
    
origin = "/home2/my-dir/TREC-CT"

for i in range(75):
    topic_no = i+1
    copy = f"TREC-CT/topic{topic_no}"
    
    qrels = qrels2lines(topic_no)
    save_qrel2files(topic_no, qrels, origin, copy, list_of_id, list_of_path)
```
